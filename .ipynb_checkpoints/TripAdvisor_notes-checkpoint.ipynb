### Step 1: Cleaning the data

```python
#Change the string total reviews to float or int
#Remove Review# data['total_reviews'] = data['total_reviews'].apply(lambda x: x.replace('review',''))
#Remove unneeded data# data['total_reviews'] = data['total_reviews'].apply(lambda x: x.replace(' s','').replace(',','').strip())
data['total_reviews'] = data['total_reviews'].apply(lambda x: float(x))
data[['total_reviews']].sample()
```

```python
#Comparing rating to total reviews
df1 = data[['title','rating','total_reviews']]
df1.sort_values(by=['rating','total_reviews'], ascending = False).sample()
```

```python
#Convert strings to float
#data['num_excellent']=data['num_excellent'].apply(lambda x: float(x.replace(',','').strip() ))
data['num_good']=data['num_good'].apply(lambda x: float(x.replace(',','').strip() ))
data['num_avg']=data['num_avg'].apply(lambda x: float(x.replace(',','').strip() ))
data['num_poor']=data['num_poor'].apply(lambda x: float(x.replace(',','').strip() ))
data['num_bad']=data['num_bad'].apply(lambda x: float(x.replace(',','').strip() ))
```

# Mapping addresses

```python
# Converting address to lat/lon marking it on a map
# !pip install geopy
import folium
from  geopy.geocoders import Nominatim

m = folium.Map(location=[40.76, -74.04], tiles="OpenStreetMap", zoom_start=12)

for address in data['address'].sample(100):
    geolocator = Nominatim()
    country = ', USA'
    loc = geolocator.geocode(address+country)
    if loc != None:
        print(address,loc.latitude,loc.longitude)
        folium.Marker([loc.latitude,loc.longitude]).add_to(m)
m
```

```python
#Store lat and lon corrd
#Run only once or you will get Error 429
import folium
from folium import plugins
from geopy.extra.rate_limiter import RateLimiter
from  geopy.geocoders import Nominatim
m = folium.Map(location=[40.76, -74.04], tiles="OpenStreetMap", zoom_start=12)
lat=[]
lon=[]
i=0
country = ', USA'
geolocator = Nominatim(timeout=5)
sgeocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

for address in data['address']:
    loc = geolocator.geocode(str(address)+country)
    if loc != None:
        lat.append(loc.latitude)
        lon.append(loc.longitude)
        #folium.CircleMarker([loc.latitude,loc.longitude], radius =1,fill_color="#3db7e4").add_to(m)
    print(i)
    i+=1
m = folium.Map(location=[40.76, -74.00], tiles="OpenStreetMap", zoom_start=12)
Sample_coord = pd.DataFrame({'lat':lat,'lon':lon}).as_matrix()

m.add_children(plugins.HeatMap(Sample_coord, radius=20))
```

```python
df_add = data[['title','address']]
```

```python
#Converts address to coord with timer to prevent 403 and 429.
from tqdm import tqdm
tqdm.pandas()
geolocator = Nominatim(timeout=5, user_agent = 'my-application')
sgeocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

df_add['location'] = df_add['address'].progress_apply(sgeocode)

pd.options.display.max_colwidth = 300
print(df_add[['location']].iloc[0,])
pd.options.display.max_colwidth = 50
```

```python
#Extract the coord from location, but loc.point provides (lat,lon, 0.0) format. ln 2 cleans up formating
df_add['point'] = df_add['location'].apply(lambda loc: list(loc.point) if loc else None)
df_add['point'] = df_add['point'].apply(lambda x: str(x).replace(', 0.0',''))
```

```python
#Create Map Markers. df_add['point'] didn't work with this method
circle = []
for coord in df_add['location']:
    try:
        create = folium.CircleMarker([coord.latitude,coord.longitude], radius=3, popup = coord, fill_color="#3db7e4" )
        circle.append(create)
    except AttributeError:
        continue
len(circle) == df_add['location'].notnull().sum()
```

```python
#Add markers to map
Manhattan = [40.76, -74.00]
m = folium.Map(location=Manhattan, tiles="OpenStreetMap", zoom_start=11)

for item in circle:
    item.add_to(m)
m
```
